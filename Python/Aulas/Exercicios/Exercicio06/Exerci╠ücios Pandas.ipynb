## Neste projeto, trabalharemos com dois arquivos csvs separados, contendo informações sobre o preço da gasolina no Brasil. 

### 1. Carregue os conjuntos de dados "gasolina_2000+.csv" e "gasolina_2010+.csv" em dois DataFrames diferentes e combine-os em um único DataFrame.

```python
import pandas as pd
gasolina_2000_df = pd.read_csv("gasolina_2000+.csv")
gasolina_2010_df = pd.read_csv("gasolina_2010+.csv")
gasolina_00_e_10_df = pd.concat([gasolina_2000_df, gasolina_2010_df])
display(gasolina_00_e_10_df)
```

### 2.Investigue as colunas e entenda o conjunto de dados usando o head() e info()

```python
display(gasolina_00_e_10_df.head())
display(gasolina_00_e_10_df.info())
```

### 3. Selecione a terceira entrada da coluna DATA INICIAL e verifique seu tipo.

```python
terceira_entrada = gasolina_00_e_10_df.iloc[2]['DATA INICIAL']
display(terceira_entrada)
print(type(terceira_entrada))
```

### 4. Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converta ambas colunas para Timestamp / Datetime.

```python
gasolina_00_e_10_df['DATA INICIAL'] = pd.to_datetime(gasolina_00_e_10_df['DATA INICIAL'], format='%Y-%m-%d')
gasolina_00_e_10_df['DATA FINAL'] = pd.to_datetime(gasolina_00_e_10_df['DATA FINAL'], format='%Y-%m-%d')
print(gasolina_00_e_10_df['DATA INICIAL'].dtype)
print(gasolina_00_e_10_df['DATA FINAL'].dtype)
```

### 5. Crie uma nova coluna para representar o ano e o mês(aaaa-mm), utilizando a coluna DATA FINAL como referência. 

```python
gasolina_00_e_10_df['ANO-MES'] = gasolina_00_e_10_df['DATA FINAL'].dt.strftime('%Y-%m')
display(gasolina_00_e_10_df)
```

### 6. Utilizando o value_counts(), liste todos os tipos de produtos contidos na base de dados.

```python
produtos = gasolina_00_e_10_df['PRODUTO'].value_counts()
display(produtos)
```

### 7. Filtre o DataFrame para obter apenas dados da 'GASOLINA COMUM'. Grave em uma nova variável.

```python
gasolina_comum_df = gasolina_00_e_10_df.loc[gasolina_00_e_10_df['PRODUTO'] == 'GASOLINA COMUM']
display(gasolina_comum_df)
```

### 8. Qual o preço médio de revenda da gasolina em agosto de 2008?

```python
gasolina_ago_2008 = gasolina_00_e_10_df.loc[gasolina_00_e_10_df['ANO-MES'] == '2008-08']
preco_medio_gasolina_ago_2008 = gasolina_ago_2008['PREÇO MÉDIO REVENDA'].mean()
display(preco_medio_gasolina_ago_2008)
```

### 9. Qual o preço médio de revenda da gasolina em maio de 2014 em São Paulo?

```python
gasolina_maio_2014 = gasolina_00_e_10_df.loc[(gasolina_00_e_10_df['ANO-MES'] == '2014-05') & (gasolina_00_e_10_df['ESTADO'] == 'SAO PAULO')]
preco_medio_gasolina_maio_2014 = gasolina_maio_2014['PREÇO MÉDIO REVENDA'].mean()
display(preco_medio_gasolina_maio_2014)
```

### 10. Você conseguiria descobrir em qual(quais) estado(s) a gasolina ultrapassou a barreira dos R$ 5,00? E quando isso ocorreu?

```python
gasolina_acima_de_5 = gasolina_comum_df[gasolina_comum_df['PREÇO MÉDIO REVENDA'] > 5]
estados_gasolina_maior_5 = gasolina_acima_de_5['ESTADO'].unique()

print(estados_gasolina_maior_5)

for estado in estados_gasolina_maior_5:
    df_estado = gasolina_acima_de_5[gasolina_acima_de_5['ESTADO'] == estado]
    print(f"\n{estado}:")
    display(df_estado[['ANO-MES','PREÇO MÉDIO REVENDA']])  
```

### 11. Qual a média de preço dos estados da região sul em 2012?

```python
preco_gasolina_comum_sul = gasolina_comum_df[gasolina_comum_df['REGIÃO'] == 'SUL']
preco_gasolina_comum_sul['ANO'] = preco_gasolina_comum_sul['DATA FINAL'].dt.year
preco_gasolina_comum_sul_2012_df = preco_gasolina_comum_sul[preco_gasolina_comum_sul['ANO'] == 2012]
print(preco_gasolina_comum_sul_2012_df['PREÇO MÉDIO REVENDA'].mean())
#display(preco_gasolina_comum_sul_2012_df)
#preco_medio_gasolina_comum_sul_2012 = PREÇO MÉDIO REVENDA 
```
