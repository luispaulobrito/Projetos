```python
import pandas as pd
import matplotlib.pyplot as plt

vendas_df = pd.read_excel("../Exercicio02/Vendas.xlsx")
produto_quantidade_df = vendas_df[['Produto', 'Quantidade']].groupby('Produto').sum()
produto_quantidade_df = produto_quantidade_df.reset_index() # redefinindo o índice
produto_quantidade_df = produto_quantidade_df.set_index("Produto") # definindo a coluna "Produto" como índice
produto_quantidade_df.plot(kind="bar", legend=None, xlabel="Produto", ylabel="Quantidade", title="Quantidade de produtos vendidos")
plt.show()
plt.savefig("produto_quantidade_df.png")
```
