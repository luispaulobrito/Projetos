```python
import pandas as pd
vendas_df = pd.read_excel("Vendas.xlsx")
display(vendas_df)
```

```python
valor_unitario_acima700 = vendas_df[(vendas_df['Valor Unitário'] > 700) & (vendas_df['Quantidade'] > 4)]
display(valor_unitario_acima700)
```

```python
produtos_df = vendas_df.groupby("Produto")
soma_df = produtos_df.sum()
mais_vendidos_df = produtos_df.max()
display(soma_df)
display(mais_vendidos_df)
```
