### Importando o pandas

```python
import pandas as pd
```

### Criando um dataframe a partir de um dicionário

```python
# dataframe = pd.DataFrame()
venda = {'data': ['15/02/2021', '16/02/2021'],
         'valor': [500, 300],
         'produto': ['feijao', 'arroz'],
         'qtde': [50, 70],
        }
vendas_df = pd.DataFrame(venda)        
```

### Visualização dos Dados
    - print
    - display

```python
display(vendas_df)
```

### Importando arquivos e bases de dados

```python
vendas_df = pd.read_excel("Vendas.xlsx")
display(vendas_df)
```

### Resumos de Visualização de Dados simples e úteis
    - head
    - shape
    - describe

```python
display(vendas_df.head(10))
print(vendas_df.shape) #mostra a quantidade de linhas da tabela
display(vendas_df.describe()) #faz um resumo da tabela, exemplo a segunda linha mostra a média
```

### Pegar 1 coluna (e os pd.Series)

```python
produtos = vendas_df[['Produto','ID Loja']]
display(produtos)
```

### .loc, um método muito importante
    - Pegar 1 linha
    - Pegar linhas de acordo com alguma condição
    - Pegar linhas e colunas específicas
    - Pegar 1 valor específico

```python
#pegar uma linha
display(vendas_df.loc[1])

#pegar 5 linhas
display(vendas_df[1:5])

#pegar linhas de acordo com uma condição
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping']
display(vendas_norteshopping_df)

#pegar várias linhas e colunas usando o loc
vendas_norteshopping_resumida_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ['Data','Produto']]
display(vendas_norteshopping_resumida_df)

#pegar um item específico
print(vendas_df.loc[1,'Produto'])
```

### Adicionar 1 coluna

```python
# adicionar coluna a partir de uma que já existe
vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.05
display(vendas_df)

#criar coluna com um valor padrão
# pode fazer assim: vendas_df['Imposto'] = 0
# Mas o ideal é fazer usando loc para o panda não reclamar
vendas_df.loc[:,'Imposto'] = 0
display(vendas_df)
```

### Adicionar 1 linha
    - Linhas de um complemento da base de dados

```python
#Observações:
#O comando de baixo eu adicionei a tabela de vendas de dezembro na tabela de Vendas, como antes eu 
# já tinha adicionado colunas as de dezembro ficaram NaN
# Se eu executar esse código novamente ele adiciona mais linhas.

vendas_dez_df = pd.read_excel("Vendas - Dez.xlsx")
display(vendas_dez_df)

vendas_df = vendas_df.append(vendas_dez_df)
display(vendas_df)
```

### Excluir linhas e colunas

```python
# axis é o eixo, coluna =1 linha =0
#vendas_df = vendas_df.drop('Imposto', axis=1)

vendas_df = vendas_df.drop(3, axis=0)
display(vendas_df)
```

### Valores Vazios
    - Deletar linhas/colunas vazias
    - Deletar linhas que possuem valores vazios
    - Preencher valores vazios (média e último valor)

```python
# deletar linhas e colunas completamente vazias
#vendas_df = vendas_df.dropna(how='all', axis=0)
#display(vendas_df)

#deletar linhas que possuem pelo menos um valor vazio
#vendas_df = vendas_df.dropna()
#display(vendas_df)

#preencher valores vazios
#preencher com a média da coluna
vendas_df['Imposto'] = vendas_df['Imposto'].fillna(vendas_df['Imposto'].mean())
display(vendas_df)

#verifica a quantidade de valores ausentes
print(vendas_df['Imposto'].isnull().sum())

#verifica a quantidade de valores ausentes independentemente do valor utilizado para representá-los
print(vendas_df['Imposto'].isna().sum())

#preencher com o valor que está acima dele
#vendas_df = vendas_df.ffill()

```

### Calcular Indicadores
    - Groupby
    - Value Counts

```python
#value counts
transacoes_loja = vendas_df['ID Loja'].value_counts()
display(transacoes_loja)

#groupby
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```

### Mesclar 2 dataframes (Procurar informações de um dataframe em outro)

```python
#li a planilha de gerentes
gerentes_df = pd.read_excel('Gerentes.xlsx')
display(gerentes_df)

#mesclei a planilha de gerentes com a planilha de vendas, como tem a coluna ID Loja nas duas ela consegue fazer isso
vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```
